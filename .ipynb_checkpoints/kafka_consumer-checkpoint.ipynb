```python
from kafka import KafkaConsumer
from json import loads

# Initialize the Kafka consumer
consumer = KafkaConsumer(
    'crypto_data',
    bootstrap_servers=['18.219.236.123:9092'],
    auto_offset_reset='latest',
    value_deserializer=lambda x: loads(x.decode('utf-8'))
)

print("Consumer started. Waiting for messages...")

try:
    for message in consumer:
        print("\nReceived message:")
        print(message.value)
except KeyboardInterrupt:
    print("\nConsumer stopped by user")
finally:
    consumer.close()
    print("Consumer closed")
```

```python
mongo_client = MongoClient('')
db = mongo_client["stock_market"]  
collection = db["realtime_data"]  

print("Connected to MongoDB and Kafka. Listening for messages...")

# Process messages from Kafka and insert into MongoDB
for message in consumer:
    data = message.value  # Deserialize the message value
    print(f"Received data: {data}")

    try:
        if isinstance(data, dict):  
            # Check for each stock in the data
            for stock, stock_data in data.items():
                timestamp = stock_data['timestamp']
                # Check if a record with the same stock and timestamp already exists
                existing_record = collection.find_one({"_id": f"{stock}_{timestamp}"})
                
                if existing_record:
                    print(f"Duplicate data for {stock} at {timestamp}. Skipping insertion.")
                else:
                    # Add a unique identifier for each stock and timestamp
                    stock_data["_id"] = f"{stock}_{timestamp}"  
                    stock_data["stock"] = stock
                    
                    # Insert the new record
                    result = collection.insert_one(stock_data)
                    print(f"Data for {stock} inserted successfully with ID: {result.inserted_id}")
        else:
            print("Invalid data format. Skipping...")
    except Exception as e:
        print(f"Error inserting data into MongoDB: {e}")
```
